```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
MOVE_AVG = 100
ALPHA = 0.4
FIGSIZE = (10,5)
```

```python
def smoothScores(scores):
    smooth = np.zeros((len(scores)-MOVE_AVG))
    for i in range(len(smooth)):
        smooth[i] = np.average(scores[i:i+100])
    return smooth
```

```python
def extendTime(scores, time):
    return np.interp(np.linspace(0, 1, time), np.linspace(0, 1, len(scores)), scores)
```

```python
with open("Atlantis-ram_500e_scores_DQN_EP.pk", "rb") as f:
    atl_ram_dqnep = smoothScores(pickle.load(f))
with open("Atlantis-ram_10000e_scores_DQN_AS.pk", "rb") as f:
    atl_ram_dqnas = smoothScores(pickle.load(f))
with open("Atlantis-ram_10000e_scores_NSTEP_QL.pk", "rb") as f:
    atl_ram_nstpql = smoothScores(pickle.load(f))

with open("BeamRider-ram_600e_scores_DQN_EP.pk", "rb") as f:
    bmr_ram_dqnep = smoothScores(pickle.load(f))
with open("BeamRider-ram_12000e_scores_DQN_AS.pk", "rb") as f:
    bmr_ram_dqnas = smoothScores(pickle.load(f))
with open("BeamRider-ram_15000e_scores_NSTEP_QL.pk", "rb") as f:
    bmr_ram_nstpql = smoothScores(pickle.load(f))

with open("Breakout-ram_2500e_scores_DQN_EP.pk", "rb") as f:
    brk_ram_dqnep = smoothScores(pickle.load(f))
with open("Breakout-ram_50000e_scores_DQN_AS.pk", "rb") as f:
    brk_ram_dqnas = smoothScores(pickle.load(f))
with open("Breakout-ram_50000e_scores_NSTEP_QL.pk", "rb") as f:
    brk_ram_nstpql = smoothScores(pickle.load(f))
    
with open("Atlantis_4000e_scores_NSTEP_QL.pk", "rb") as f:
    atl_nstpql = smoothScores(pickle.load(f))
with open("BeamRider_4000e_scores_NSTEP_QL.pk", "rb") as f:
    bmr_nstpql = smoothScores(pickle.load(f))
with open("Breakout_15000e_scores_NSTEP_QL.pk", "rb") as f:
    brk_nstpql = smoothScores(pickle.load(f))
```

```python
ext_atl_ram_dqnep = extendTime(atl_ram_dqnep, 5087)
ext_atl_ram_dqnas = extendTime(atl_ram_dqnas, 4388)
ext_atl_ram_nstpql = extendTime(atl_ram_nstpql, 4000)

ext_bmr_ram_dqnep = extendTime(bmr_ram_dqnep, 5981)
ext_bmr_ram_dqnas = extendTime(bmr_ram_dqnas, 5183)
ext_bmr_ram_nstpql = extendTime(bmr_ram_nstpql, 6825)

ext_brk_ram_dqnep = extendTime(brk_ram_dqnep, 4349)
ext_brk_ram_dqnas = extendTime(brk_ram_dqnas, 3592)
ext_brk_ram_nstpql = extendTime(brk_ram_nstpql, 4820)

ext_atl_nstpql = extendTime(atl_nstpql, 4283)
ext_bmr_nstpql = extendTime(bmr_nstpql, 6033)
ext_brk_nstpql = extendTime(brk_nstpql, 6818)
```

```python
plt.figure(figsize=FIGSIZE)
plt.plot(ext_atl_ram_dqnep, label="DQN_EP_RAM", alpha=ALPHA)
plt.plot(ext_atl_ram_dqnas, label="DQN_AS_RAM", alpha=ALPHA)
plt.plot(ext_atl_ram_nstpql, label="NSTEP_QL_RAM", alpha=ALPHA)
plt.plot(ext_atl_nstpql, label="NSTEP_QL_SCREEN", alpha=ALPHA)
plt.title("Atlantis, Training Score versus Time")
plt.legend()
```

```python
plt.figure(figsize=FIGSIZE)
plt.plot(ext_bmr_ram_dqnep, label="DQN_EP_RAM", alpha=ALPHA)
plt.plot(ext_bmr_ram_dqnas, label="DQN_AS_RAM", alpha=ALPHA)
plt.plot(ext_bmr_ram_nstpql, label="NSTEP_QL_RAM", alpha=ALPHA)
plt.plot(ext_bmr_nstpql, label="NSTEP_QL_SCREEN", alpha=ALPHA)
plt.title("BeamRider, Training Score versus Time")
plt.legend()
```

```python
plt.figure(figsize=FIGSIZE)
plt.plot(ext_brk_ram_dqnep, label="DQN_EP_RAM", alpha=ALPHA)
plt.plot(ext_brk_ram_dqnas, label="DQN_AS_RAM", alpha=ALPHA)
plt.plot(ext_brk_ram_nstpql, label="NSTEP_QL_RAM", alpha=ALPHA)
plt.plot(ext_brk_nstpql, label="NSTEP_QL_SCREEN", alpha=ALPHA)
plt.title("Breakout, Training Score versus Time")
plt.legend()
```
